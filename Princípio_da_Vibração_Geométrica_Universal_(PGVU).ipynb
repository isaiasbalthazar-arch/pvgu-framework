```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =============================
# CONFIGURAÇÃO GERAL
# =============================

N = 10000  # Universos simulados (rápido e estável)

np.random.seed(42)

# =============================
# MÓDULO 1 — RAZÃO HARMÔNICA UNIVERSAL
# =============================

# Densidade do vácuo simulada (normalizada)
rho_vacuo = np.random.lognormal(mean=0, sigma=0.4, size=N)

# Frequência geométrica fundamental
f_geom = np.random.lognormal(mean=0, sigma=0.4, size=N)

# Razão Harmônica Universal
RHU = f_geom / (rho_vacuo ** 0.25)

# =============================
# MÓDULO 2 — TOPOLOGIA SIMPLIFICADA
# =============================

# Parâmetro de curvatura (k)
curvatura = np.random.normal(0, 0.5, N)

# Índice topológico simplificado
indice_topologico = np.exp(-np.abs(curvatura))

# =============================
# MÓDULO 3 — DUALIDADE QUÂNTICA
# =============================

# Grau de coerência quântica
coerencia = np.random.uniform(0, 1, N)

# Ruído de vácuo
ruido_vacuo = np.random.normal(0, 0.15, N)

# Métrica de dualidade (interferência residual)
dualidade = coerencia * np.exp(-np.abs(ruido_vacuo))

# =============================
# CRITÉRIO DE UNIVERSO ESTÁVEL (PNO OPERACIONAL)
# =============================

estavel = (
    (RHU > np.percentile(RHU, 40)) &
    (RHU < np.percentile(RHU, 60)) &
    (indice_topologico > 0.6) &
    (dualidade > 0.5)
)

frac_estaveis = estavel.sum() / N * 100

# =============================
# EXPORTAR RESULTADOS
# =============================

df = pd.DataFrame({
    "rho_vacuo": rho_vacuo,
    "f_geom": f_geom,
    "RHU": RHU,
    "curvatura": curvatura,
    "indice_topologico": indice_topologico,
    "dualidade": dualidade,
    "universo_estavel": estavel
})

df.to_csv("paisagem_multiversal_LPVM_fast.csv", index=False)

# =============================
# VISUALIZAÇÕES
# =============================

plt.figure(figsize=(7,5))
plt.hist(RHU, bins=100)
plt.title("Distribuição da Razão Harmônica Universal (RHU)")
plt.xlabel("RHU")
plt.ylabel("Frequência")
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(indice_topologico, dualidade, s=2)
plt.xlabel("Índice Topológico")
plt.ylabel("Dualidade Quântica")
plt.title("Correlação Geometria vs Dualidade")
plt.show()

print("====================================")
print("LABORATÓRIO MULTIVERSAL — RESULTADOS")
print("====================================")
print(f"Universos simulados: {N}")
print(f"Universos estruturalmente estáveis: {estavel.sum()}")
print(f"Fraçao de estabilidade vibracional: {frac_estaveis:.2f}%")
print("Arquivo salvo: paisagem_multiversal_LPVM_fast.csv")
```

```python
# ============================================
# LABORATÓRIO MULTIVERSAL — LPVM v2
# Autor conceitual: Isaías Balthazar da Silva
# Implementação experimental científica
# ============================================

import numpy as np
import pandas as pd

# -------------------------
# CONFIGURAÇÕES GERAIS
# -------------------------

NUM_UNIVERSOS = 10000
NUM_RODADAS = 5

np.random.seed(None)

# -------------------------
# PARÂMETROS FÍSICOS SIMULADOS
# -------------------------

# Constante de estrutura fina (simulada)
ALPHA_MIN = 1/200
ALPHA_MAX = 1/80

# Frequência vibracional geométrica
VIB_MIN = 0.1
VIB_MAX = 10.0

# Ruído quântico
NOISE_MIN = 0.0
NOISE_MAX = 1.0

# -------------------------
# CLASSIFICAÇÃO DE UNIVERSOS
# -------------------------

def classificar_universo(vib, ruido):
    if vib < 3.0 and ruido < 0.3:
        return "Campo"
    elif vib < 7.0 and ruido < 0.7:
        return "Brana"
    else:
        return "Caotico"

# -------------------------
# MÉTRICA DE ESTABILIDADE VIBRACIONAL
# -------------------------

def estabilidade_vibracional(alpha, vib, ruido):

    # Harmonia geométrica (ressonância)
    ressonancia = np.exp(-abs(alpha - (1/137)))

    # Coerência vibracional
    coerencia = np.exp(-ruido) * np.cos(vib * np.pi / 10)

    # Índice estrutural combinado
    indice = ressonancia * coerencia

    return indice

# -------------------------
# CRITÉRIO FÍSICO DE ESTABILIDADE
# -------------------------

LIMIAR_ESTABILIDADE = 0.35

# -------------------------
# EXECUÇÃO MULTIVERSAL
# -------------------------

resultados_globais = []

for rodada in range(NUM_RODADAS):

    alphas = np.random.uniform(ALPHA_MIN, ALPHA_MAX, NUM_UNIVERSOS)
    vibracoes = np.random.uniform(VIB_MIN, VIB_MAX, NUM_UNIVERSOS)
    ruidos = np.random.uniform(NOISE_MIN, NOISE_MAX, NUM_UNIVERSOS)

    estaveis = 0
    dados = []

    for i in range(NUM_UNIVERSOS):

        classe = classificar_universo(vibracoes[i], ruidos[i])
        indice = estabilidade_vibracional(alphas[i], vibracoes[i], ruidos[i])

        status = "Estavel" if indice > LIMIAR_ESTABILIDADE else "Instavel"

        if status == "Estavel":
            estaveis += 1

        dados.append([
            rodada+1,
            alphas[i],
            vibracoes[i],
            ruidos[i],
            classe,
            indice,
            status
        ])

    fracao = estaveis / NUM_UNIVERSOS * 100

    print(f"Rodada {rodada+1}: Universos estáveis = {estaveis} ({fracao:.2f}%)")

    resultados_globais.append(fracao)

# -------------------------
# ANÁLISE ESTATÍSTICA FINAL
# -------------------------

media = np.mean(resultados_globais)
desvio = np.std(resultados_globais)

print("\n====================================")
print("RESULTADO GLOBAL — PAISAGEM MULTIVERSAL")
print("====================================")
print(f"Média estabilidade vibracional: {media:.2f}%")
print(f"Desvio padrão: {desvio:.2f}%")

# -------------------------
# EXPORTAÇÃO CIENTÍFICA
# -------------------------

df = pd.DataFrame(dados, columns=[
    "Rodada",
    "Constante_Estrutura_Fina",
    "Vibracao_Geometrica",
    "Ruido_Quantico",
    "Classe_Universo",
    "Indice_Estabilidade",
    "Status"
])

arquivo = "paisagem_multiversal_LPVM_public_release.csv"
df.to_csv(arquivo, index=False)

print(f"\nArquivo científico salvo: {arquivo}")
```
